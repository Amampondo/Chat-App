```python
# import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import warnings
warnings.filterwarnings("ignore")

# Load the dataset
df = pd.read_csv('/cloudData.csv')
# we will be selecting usage_end_time', 'cost', 'service_type', 'region', 'usage_amount','usage_unit
df = df[['usage_end_time', 'cost', 'service_type', 'region', 'usage_amount','usage_unit']]

# Select only cost for the isolation forest model test
df_cost = df[['cost',]]
anomaly_inputs = ['cost', 'usage_amount']

# Instantiate the model and fit the data to it
model=IsolationForest(n_estimators=1000,max_samples='auto', contamination=float(0.1),max_features=1.0,random_state=0)
model.fit(df[anomaly_inputs])

# Get the score and anomaly flag
df['scores']=model.decision_function(df[anomaly_inputs])
df['anomaly']=model.predict(df[anomaly_inputs])
# view the data
df.sort_values(by=['cost'])

# fetch all the anomalies
anomaly=df.loc[df['anomaly']==-1]
anomaly.head(40).sort_values(by=['cost'])

#df_merged = pd.merge(df, df_cost, on=["cost"])
#df_merged.sort_values(by=['cost'])
```
